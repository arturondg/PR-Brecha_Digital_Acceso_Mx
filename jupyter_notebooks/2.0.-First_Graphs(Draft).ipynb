```python
import pandas as pd
```

```python
path_hogar = ("../data/ENDUTIH_2021_transformed/ENDUTIH_HOGAR_2021.csv")
df_hogar = pd.read_csv(path_hogar)
df_hogar.head()
```

```python
df_e = df_hogar.loc[0:1000:20,["UPM", "P4_1_6", "FAC_HOG", "UPM_DIS", "EST_DIS", "DOMINIO", "TLOC"]]
df_e
```

```python
total_poblacion = sum(df_hogar['FAC_HOG'])
#df_hogar["FAC_HOG"].sum()
total_poblacion
```

```python
X = (df_e['FAC_HOG'] * df_e['UPM_DIS']).sum() + (df_e.loc[2, 'FAC_HOG'] * df_e.loc[2, 'UPM_DIS'] * df_e.loc[2, 'TLOC'])
X 
```

```python
# Calcular el total de la población
total_poblacion = sum(df_hogar['FAC_HOG'])

# Seleccionar las UPM_DIS
upm_dis = df_hogar['UPM_DIS']

# Calcular los pesos de expansión para cada UPM_DIS
pesos_expansion = df_hogar['FAC_HOG'] / sum(df_hogar['FAC_HOG'])

# Calcular la expansión de la variable de interés para cada UPM_DIS
expansion_p4_1_6 = upm_dis.map(dict(zip(upm_dis.unique(), df_hogar['P4_1_6'] * pesos_expansion)))

# Calcular la estimación del total de la variable de interés en la población
estimacion_total = expansion_p4_1_6.sum() / pesos_expansion.sum()

print('Estimación del total de hogares con celulares en la población:', round(estimacion_total))
```

```python
df_pc_cel_hog = df_hogar[["ESTRATO", "P4_1_6", "P4_2_1", "P4_2_2"]]
df_pc_cel_hog.head()
```

```python
df_pc_cel_hog = df_pc_cel_hog.rename(columns={"P4_1_6" : "Celular", "P4_2_1" : "PC", "P4_2_2" : "Portatil" })
df_pc_cel_hog.head()
```

```python
for column in df_pc_cel_hog.columns.values.tolist():
    print(column)
    print(df_pc_cel_hog[column].value_counts())
```

```python
df_pc_cel_hog["ESTRATO"] = df_pc_cel_hog["ESTRATO"].astype("object")
df_pc_cel_hog["ESTRATO"].value_counts()
```

```python
print(df_pc_cel_hog["ESTRATO"].dtypes)
```

```python
df_pc_cel_hog["ESTRATO"] = df_pc_cel_hog["ESTRATO"].replace([1, 2, 3, 4], ["Bajo", "Medio Bajo", "Medio Alto", "Alto"])
```

```python
df_pc_cel_hog["ESTRATO"].value_counts()
```

```python
df_pc_cel_hog2 = df_pc_cel_hog.replace(2, 0)
for column in df_pc_cel_hog2.columns.values.tolist():
    print(column)
    print(df_pc_cel_hog2[column].value_counts())
```

```python
df_pc_cel_hog2.groupby("ESTRATO").sum()
```

```python
df_pc = df_pc_cel_hog2[["ESTRATO","PC", "Portatil"]]
df_pc.head()
```

```python
df_compu = (df_pc["PC"] + df_pc["Portatil"]).apply(lambda x: 1 if x >= 1 else 0)
df_compu2 = pd.concat([df_pc, df_compu ], axis=1)
df_compu2.head()
```
